## Checking Distribution of Dependent Variables
I have 2 variables that I am interested in explaining: performance on the task as well as the time it took to complete a trial. It's important to visualize the distribution of these variables, as well as test for normality. 

```python
#stats
def check_normality(data):
    D, p = stats.kstest(data, 'norm')
    print 'KS stat: %.3f - p-value: %.3f' %(D,p)
    stat, p = shapiro(data)
    print 'Shapiro-Wilk Test: p-value: %.3f' %p
    stat, p = normaltest(data)
    print 'Agostino k^2 test: p-value: %.3f' %p
    print 'Anderson test:'
    result = anderson(data)
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < result.critical_values[i]:
            print('%.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
        else:
            print('%.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))

#create figure            
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14,10))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=0.2, hspace=0.4)
#histogram
plt.axes(axes.flatten()[0])
DSR['Perf'].plot.hist(bins=11)
axes.flatten()[0].set_title('DSR Performance - Trained Animals', fontsize=16)
axes.flatten()[0].set_xlabel('Accuracy')
#qqplot
qqplot(DSR['Perf'], line='s', ax = axes.flatten()[1])
axes.flatten()[1].set_title('DSR Performance - Trained Animals', fontsize=16)


#histogram  
plt.axes(axes.flatten()[2])     
PSR['Perf'].plot.hist(bins=11)
axes.flatten()[2].set_title('PSR Performance - Trained Animals', fontsize=16)
axes.flatten()[2].set_xlabel('Accuracy')
#qqplot
qqplot(PSR['Perf'], line='s', ax = axes.flatten()[3])
axes.flatten()[3].set_title('PSR Performance - Trained Animals', fontsize=16)


#histogram
plt.axes(axes.flatten()[4])     
ax = Naive['Perf'].plot.hist(bins=8)
axes.flatten()[4].set_title('DSR Performance - Naive Animals', fontsize=16)
axes.flatten()[4].set_xlabel('Accuracy')
#qqplot
qqplot(Naive['Perf'], line='s', ax = axes.flatten()[5])
axes.flatten()[5].set_title('DSR Performance - Naive Animals', fontsize=16)
#plt.savefig('Performance_Normality.jpg', dpi = 500, bbox_inches = 'tight')
plt.show()

print '*' * 80
print 'DSR - TRAINED'
check_normality(DSR['Perf'])
print '*' * 80
print 'PSR - TRAINED'
check_normality(PSR['Perf'])
print '*' * 80
print 'DSR - Naive'
check_normality(Naive['Perf'])
```

## Saline Comparisons
I am first interested in checking whether, for any of the datasets, there were differences among saline infusions. Statsmodels makes that really easy with the use of R-like patsy formulas. Differences in performance between saline datasets will also include information about block, and animal. The Naive dataset only has one saline session per subject, so we can't check for that there.

```python
#should probably be a repeated-measures 2-way ANOVA which is similar to a mixed linear model
formula = 'Perf ~ Order + C(Notes, Treatment(reference="PL-saline")) * C(Rev1)'
saline_DSR = DSR[[w.find('saline')>=0 for w in DSR['Infusion_String']]]
saline_PSR = PSR[[w.find('saline')>=0 for w in PSR['Infusion_String']]]

saline_DSR_LMM = smf.mixedlm(formula, data = saline_DSR, groups = saline_DSR['Animal']).fit()
saline_PSR_LMM = smf.mixedlm(formula, data = saline_PSR, groups = saline_PSR['Animal']).fit()

print '{0:{fill}{align}80}'.format('Saline DSR',fill='-', align='^')
display(saline_DSR_LMM.summary())

print '{0:{fill}{align}80}'.format('Saline PSR',fill='-', align='^')
display(saline_PSR_LMM.summary())
```

## No Differences in Saline Infusions
Since there is no difference in performance between saline groups, we will from now on combine those sessions together and refer to them as the same.

```python
#combining saline into all one label - in a new column
DSR['Infusion_String_Reduced'] = DSR['Infusion_String']
DSR.loc[[ w.find('saline') >= 0 for w in DSR['Infusion_String_Reduced']],'Infusion_String_Reduced'] = 'saline'
PSR['Infusion_String_Reduced'] = PSR['Infusion_String']
PSR.loc[[ w.find('saline') >= 0 for w in PSR['Infusion_String_Reduced']],'Infusion_String_Reduced'] = 'saline'
```

## Effects of PFC Inactivation on Performance
Since there were no differences in performance between the saline groups (above), we will pool them together for the rest of the analyses. There were some differences in timing, however, which we will report but will not affect the way in which we go ahead with the analysis.

```python
performance_DSR = smf.mixedlm('Perf ~ C(Order) + \
                           C(Infusion_String_Reduced, Treatment(reference="saline")) * \
                           C(Rev1)',
                           data = DSR,
                           groups = DSR['Animal']).fit()

display(performance_DSR.summary())
```

## Results: DSR
The results boil down to this:
effects of CONTRA X REV1 - p = 0.009
effects of mPFC X REV1 - p = 0.000

I guess we already knew this much, but not while controling for everything as well as we did. I wondered whether the effects of contralateral inactivation were due to having just 1 mPFC but I guess it's not because of the ipsilateral results. Let's do the same for PSR, with the added Uncertainty term.

```python
performance_PSR = smf.mixedlm('Errors ~ C(Order) + \
                           Uncertainty * C(Infusion_String_Reduced, Treatment(reference="saline")) * \
                           C(Rev1)',
                           data = PSR,
                           groups = PSR['Animal']).fit()

display(performance_PSR.summary())
```

## Results: PSR
The results boil down to this:
effects of Rev1 - p = 0.003
effects of mPFC X REV1 - p = 0.049
effects of Uncertainty - p = 0.01
effects of Uncertainty X mPFC X Rev1 - p = 0.007

The effects of Rev1 and Uncertainty tell us more about the tasks than about animals behaving differently under different types of PFC inactivation. The performance on the reversal is worse than on the first block by about ~0.05. That makes sense to me. The higher the uncertainty, the worse animals do, which also makes sense, and this is true for all treatments. The other 2 results are specific to mPFC inactivations. Performance on the reversal is lower. There is a "trend" towards CONTRA X REV1 - p = 0.078. The last one is definitely the weirdest because of the direction of the coefficient, i.e. it is positive. It reads, the more uncertainty there was during the reversal when animals had an inactivated mPFC, the *better* they did. 

```python
performance_Naive = smf.mixedlm('Errors ~ C(Order) * \
                               C(Infusion_String, Treatment(reference="Saline-Naive")) * \
                               C(Rev1)',
                               data = Naive,
                               groups = Naive['Animal']).fit()


display(performance_Naive.summary())
```

## Results: Naive
The results boil down to this:
effects of Rev1 - p = 0.000

There is a huge effect of Rev1, which is to be expected, and it was true across treatments. Other than that, there is no clear effect of inactivation. 

## Post Analysis
We are also interested in reaction time and any goal biases that may arise. Below is the same type of analysis we've been doing (Linear Mixed Effects Model), but this time we have reaction time as the dependent variable. 

```python
timing_formula = 'Timing ~  C(Infusion_String_Reduced, Treatment(reference="saline")) * C(Rev1)'
timing_formula_Naive = 'Timing ~ C(Infusion_String, Treatment(reference="Saline-Naive")) * C(Rev1)'
timing_DSR = smf.mixedlm(timing_formula, data = DSR, groups = DSR['Animal']).fit()
timing_PSR = smf.mixedlm(timing_formula, data = PSR, groups = PSR['Animal']).fit()
timing_Naive = smf.mixedlm(timing_formula_Naive, data = Naive, groups = Naive['Animal']).fit()


print '{0:{fill}{align}80}'.format('Timing DSR',fill='-', align='^')
display(timing_DSR.summary())

print '{0:{fill}{align}80}'.format('Timing PSR',fill='-', align='^')
display(timing_PSR.summary())

print '{0:{fill}{align}80}'.format('Timing Naive',fill='-', align='^')
display(timing_Naive.summary())
```

```python
goal_bias_DSR = smf.mixedlm('Perf ~ C(GA)', data = DSR, groups = DSR['Animal']).fit()
goal_bias_PSR = smf.mixedlm('Perf ~ C(GA)', data = PSR, groups = PSR['Animal']).fit()
goal_bias_Naive = smf.mixedlm('Perf ~ C(GA)', data = Naive, groups = Naive['Animal']).fit()


print '{0:{fill}{align}80}'.format('Goal Bias DSR',fill='-', align='^')
display(goal_bias_DSR.summary())
print '{0:{fill}{align}80}'.format('Goal Bias PSR',fill='-', align='^')
display(goal_bias_PSR.summary())
print '{0:{fill}{align}80}'.format('Goal Bias Naive',fill='-', align='^')
display(goal_bias_Naive.summary())
```

## Conclusions On Timing and Goal Bias

There is definitely some differences in timing. In the DSR-Trained dataset, there is an effect of Reversal, so that after the reversal they go faster. Also, when they received infusions of muscimol into the mPFC, Ipsilateral, and Contralateral, they went faster. For the PSR-Trained dataset, the same pattern holds, except there was no effect of mPFC but a strong trend (p=0.056). For the Naive dataset, however, there was no difference. 

There also appears to be some bias of goal, so that when blocks where the East was the high-payout goal, animals tended to do better. This was true for the DSR-Trained and PSR-Trained datasets. This was not true for the Naive dataset.
